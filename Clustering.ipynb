```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Load the CSV file into a DataFrame
# Encode data for event_type field
label_encoder = LabelEncoder()
df['event_type_encoded'] = label_encoder.fit_transform(df['event_type'])
df
```

```python
# Encoder data for category_code field by One-Hot Encoding
one_hot_encoded_df = pd.get_dummies(df['category_code'], prefix='category', dummy_na=True)
df = pd.concat([df, one_hot_encoded_df], axis=1)
df
```

```python
one_hot_columns = one_hot_encoded_df.columns.tolist()
one_hot_columns
```

```python
# Group by user_id for user-level aggregation
user_features = df.groupby('user_id').agg(
    event_time=('event_time', 'max'),             # Last interaction time
    activity_frequency=('event_type', 'count'),  # Frequency of activity
    unique_products=('product_id', 'nunique'),   # Number of unique products
    unique_brands=('brand', 'nunique'),          # Number of unique brands
    average_spend=('price', 'mean'),             # Average spending
    most_common_category=('category_code', lambda x: x.mode()[0] if not x.mode().empty else None),  # Most common category
).reset_index()

# Calculate recency
current_time = df['event_time'].max()
user_features['recency'] = (current_time - user_features['event_time']).dt.total_seconds()

# Aggregate event_type_encoded counts (if needed for deeper analysis)
event_counts = df.groupby(['user_id', 'event_type_encoded']).size().unstack(fill_value=0).reset_index()
event_counts.columns.name = None  # Remove the name from columns

# Merge event_type_encoded counts back into user_features
user_data = pd.merge(user_features, event_counts, on='user_id', how='left')

# Display final user-level dataset
user_data
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Define the final features to scale (including one-hot encoded columns)
features_to_scale = [
    'activity_frequency',  # Total events
    'unique_products',     # Number of unique products
    'unique_brands',       # Number of unique brands
    'average_spend',       # Average spend
    'recency',             # Time since last interaction
]

# Drop rows with missing values in the selected features
user_features_cleaned = user_features.dropna(subset=features_to_scale)

# Standardize the selected features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(user_features_cleaned[features_to_scale])

# Verify the shape and sample of the scaled data
print(f"Scaled data shape: {scaled_data.shape}")
scaled_data
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Function to calculate SSE (Sum of Squared Errors) for different numbers of clusters
sse = []
K_range = range(1, 10)  # Testing cluster counts from 1 to 10

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    sse.append(kmeans.inertia_)  # inertia_ represents the total SSE for the current number of clusters

# Plot the Elbow graph
plt.figure(figsize=(10, 6))
plt.plot(K_range, sse, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('SSE')
plt.title('Elbow Method to Determine Optimal Number of Clusters')
plt.show()
```

```python
# Select the optimal number of clusters, e.g., k = 5 from the Elbow graph
optimal_k = 5

# Apply K-means with the optimal number of clusters
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
user_data['cluster'] = kmeans.fit_predict(scaled_data)

# Display the user data with assigned cluster labels
user_data
```

Here’s a concise summary of the **pros and cons** of the four clustering models discussed:

---

### **1. K-Means**
#### **Pros**:
- **Fast and scalable**: Works well on large datasets with lower dimensionality.
- **Simple and interpretable**: Easy to understand and implement.
- **Clear output**: Produces distinct, non-overlapping clusters.

#### **Cons**:
- **Assumes spherical clusters**: Poor performance on non-spherical or unevenly sized clusters.
- **Sensitive to outliers**: Outliers can skew cluster centers.
- **Fixed `k`**: Requires predefining the number of clusters.

---

### **2. DBSCAN**
#### **Pros**:
- **Handles noise**: Identifies and labels outliers as noise.
- **Flexible cluster shapes**: Finds arbitrarily shaped clusters.
- **No `k` parameter**: Automatically determines the number of clusters.

#### **Cons**:
- **Parameter sensitivity**: Performance heavily depends on `eps` and `min_samples`.
- **Not scalable**: Can be slow on large datasets (\(O(n^2)\)).
- **Sparse data issues**: Struggles with high-dimensional datasets.

---

### **3. Agglomerative Clustering**
#### **Pros**:
- **No need to predefine clusters (`k`)**: Can stop at any desired number of clusters.
- **Hierarchical structure**: Provides a dendrogram for hierarchical relationships.
- **Handles arbitrary shapes**: Works well for various cluster structures.

#### **Cons**:
- **Computationally expensive**: Slower on large datasets (\(O(n^2)\) for distance matrix).
- **Sensitive to noise**: Outliers can distort cluster hierarchy.
- **Less scalable**: Not suitable for very large datasets.

---

### **4. Gaussian Mixture Model (GMM)**
#### **Pros**:
- **Probabilistic approach**: Assigns soft cluster memberships (probabilities).
- **Flexible shapes**: Can model non-spherical clusters.
- **Mathematically robust**: Captures complex relationships in data.

#### **Cons**:
- **Sensitive to initialization**: Poor initialization may lead to suboptimal clusters.
- **Requires `k`**: Needs the number of clusters to be predefined.
- **Computationally expensive**: Slower than K-Means for large datasets.

---

### **When to Use Each Model**
| **Scenario**                          | **Best Model**           |
|---------------------------------------|--------------------------|
| Large dataset with simple clusters    | **K-Means**              |
| Clusters with noise or outliers       | **DBSCAN**               |
| Hierarchical relationships needed     | **Agglomerative**        |
| Overlapping, soft clusters            | **GMM**                  |

```python
from sklearn.cluster import DBSCAN, AgglomerativeClustering
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


# Function to evaluate clustering methods
def evaluate_clustering(labels, data):
    if len(np.unique(labels)) > 1:  # Silhouette score needs at least 2 clusters
        score = silhouette_score(data, labels)
        print(f"Silhouette Score: {score:.4f}")
    else:
        print("Silhouette Score: Not applicable (only one cluster identified).")
    return labels

# DBSCAN
print("=== DBSCAN ===")
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(scaled_data)
user_data['DBSCAN_cluster'] = dbscan_labels
evaluate_clustering(dbscan_labels, scaled_data)
```

```python
# Agglomerative Clustering
print("\n=== Agglomerative Clustering ===")
agglo = AgglomerativeClustering(n_clusters=4)
agglo_labels = agglo.fit_predict(scaled_data)
user_data['Agglomerative_cluster'] = agglo_labels
evaluate_clustering(agglo_labels, scaled_data)
```

```python
# Gaussian Mixture Model
print("\n=== Gaussian Mixture Model ===")
gmm = GaussianMixture(n_components=4, random_state=42)
gmm_labels = gmm.fit_predict(scaled_data)
user_data['GMM_cluster'] = gmm_labels
evaluate_clustering(gmm_labels, scaled_data)
```

```python
# Function to analyze and describe cluster behaviors
def analyze_clusters(data, cluster_col, method_name):
    print(f"\n=== {method_name} Clustering Analysis ===")
    # Initialize a list to store the analysis results
    analysis_results = []

    # Group data by the clustering labels
    grouped = data.groupby(cluster_col)

    for cluster, group in grouped:
        # Ignore noise (-1 cluster) for DBSCAN
        if cluster == -1:
            print(f"\nCluster {cluster} (Noise):")
            print(f"Number of users in the cluster: {len(group)}")
            print("Behavior: Noise points (outliers). No specific behavioral insights.")
            continue

        # Key metrics for the cluster
        unique_users = group['user_id'].nunique()
        avg_activity = group['activity_frequency'].mean()
        avg_spend = group['average_spend'].mean()
        avg_unique_products = group['unique_products'].mean()
        avg_unique_brands = group['unique_brands'].mean()
        avg_recency_days = (group['recency'].mean() / 86400)  # Convert recency to days

        # Determine behavior
        behavior = ""
        strategy = ""
        if avg_spend > 500 and avg_activity < 10:
            behavior = "High spenders with low activity - likely to purchase high-value items infrequently."
            strategy = "Focus on personalized high-end product recommendations, exclusive offers, and premium services."
        elif avg_activity <= 10 and avg_spend <= 300:
            behavior = "Price-sensitive, low engagement shoppers - make fewer purchases and spend less."
            strategy = "Provide small discounts, bundled deals, and emphasize affordability to encourage engagement."
        elif avg_activity > 200 and avg_unique_products > 50:
            behavior = "Highly active explorers - frequently browse and interact with many products and brands."
            strategy = "Leverage personalized recommendations, interactive content (reviews, comparisons), and exploration-focused campaigns."
        elif 50 < avg_activity <= 200 and avg_spend > 200:
            behavior = "Moderately active, with balanced product and brand interactions."
            strategy = "Strengthen loyalty with targeted promotions, rewards programs, and emphasize consistency across familiar brands."
        elif avg_recency_days < 30:
            behavior = "Recently active users with potential for higher future engagement."
            strategy = "Reinforce engagement with follow-ups, limited-time offers, and timely personalized messages."
        else:
            behavior = "Mixed or average behavior patterns."
            strategy = "Use general engagement strategies like loyalty programs and personalized offers to maintain activity."

        # Additional insights
        additional_insight = ""
        if avg_unique_brands > 20:
            additional_insight += "Users show interest in a wide variety of brands, indicating they value comparison shopping. "
        if avg_unique_products < 5:
            additional_insight += "Users have narrow interests, focusing on a limited range of products. "
        if avg_activity > 100 and avg_spend < 200:
            additional_insight += "High engagement but low spending, suggesting interest in exploration without frequent purchases. "
        if avg_recency_days > 90:
            additional_insight += "Users haven't interacted recently, suggesting churn risks."

        # Append results
        analysis_results.append({
            'Cluster': cluster,
            'Number of Users': unique_users,
            'Average Activity Frequency': avg_activity,
            'Average Unique Products': avg_unique_products,
            'Average Spending': avg_spend,
            'Average Unique Brands': avg_unique_brands,
            'Average Recency (Days)': avg_recency_days,
            'Behavior': behavior,
            'Strategy': strategy,
            'Additional Insight': additional_insight.strip()
        })

    # Convert results to a DataFrame
    analysis_df = pd.DataFrame(analysis_results)

    # Print or save results
    excel_file_path = f'{method_name}_cluster_analysis.xlsx'
    analysis_df.to_excel(excel_file_path, index=False, engine='openpyxl')
    print(f"Cluster analysis for {method_name} saved to {excel_file_path}")

# Analyze clusters for each method
analyze_clusters(user_data, 'DBSCAN_cluster', 'DBSCAN')
analyze_clusters(user_data, 'Agglomerative_cluster', 'Agglomerative')
analyze_clusters(user_data, 'GMM_cluster', 'Gaussian Mixture Model')
```
